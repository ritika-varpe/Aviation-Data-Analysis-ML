Encoding the Categorical Data

```python
X=df
# Categorical boolean mask
categorical_feature_mask = X.dtypes==object
# filter categorical columns using mask and turn it into a list
categorical_cols = X.columns[categorical_feature_mask].tolist()
```

```python
# import labelencoder
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
le = LabelEncoder()
```

```python
# apply le on categorical feature columns
X[categorical_cols] = X[categorical_cols].apply(lambda col: le.fit_transform(col))
X[categorical_cols].head(10)
```

clustering 

```python
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.cluster import KMeans
import pandas as pd
```

```python
df_new=pd.get_dummies(df)
df_new.head()
```

```python
df_kmeans=pd.get_dummies(df,drop_first=True)
df_kmeans.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_df_kmeans=scaler.fit_transform(df_kmeans)
```

```python
from sklearn.cluster import KMeans
kmeans_model=KMeans(n_clusters=3)
clusters=kmeans_model.fit_predict(df_kmeans)
df_kmeans.insert(df_kmeans.columns.get_loc("category"),"Clusters",clusters)
df_kmeans
```

```python
df_kmeans['Clusters'].value_counts()
```

```python
df_kmeans.Clusters.unique()
```

```python
ssd=[]
for k in range(2,9):
  kmeans_model=KMeans(n_clusters=k)
  kmeans_model.fit(df_kmeans)
  ssd.append(kmeans_model.inertia_)
plt.figure(figsize=(6,4),dpi=101)
plt.plot(range(2,9),ssd,color='green',marker='o')
plt.xlabel("number of clusters(K)")
plt.ylabel("SSD for X")
plt.show()
```

```python
df1=df[['totalReceived','activeGrievancesWithoutEscalation','grievancesAdditionalInfoNotProvided','grievancesWithoutFeedback','category']]
df1.head()
```

```python
X=df1.iloc[:,:-1]
feature_cols=['totalReceived','activeGrievancesWithoutEscalation','grievancesAdditionalInfoNotProvided','grievancesWithoutFeedback']
Y=df1.iloc[:,4].values
```

```python
from sklearn.tree import DecisionTreeClassifier
regressor = DecisionTreeClassifier(criterion='gini',max_depth=4,splitter='best')
X
regressor.fit(X,Y)

X_in=np.array([2,0,1,1])
y_pred=regressor.predict([X_in])
print("Prediction: ",y_pred)
```
